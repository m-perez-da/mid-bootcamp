```python
import pandas as pd
import numpy as np
```

```python
# read quotes file (containing data for all companies listed in american markets, or al least it is supposed)
quotes_all = pd.read_csv('american_markets_quotes.csv')
```

```python
quotes_all
```

```python
quotes_all.dtypes
```

```python
quotes_all.isna().sum()
```

```python
quotes_all = quotes_all.drop(['high_price','low_price'], axis=1)  # useless columns for analysis
```

```python
quotes_all['symbol'].nunique()  # number of listed companies
```

```python
sp500_companies= pd.read_csv('sp500_companies.csv')  # DataFrame with list of S&P500 companies and sector
```

```python
sp500_companies
```

```python
sp500_companies = sp500_companies.drop(['SUB-INDUSTRY'], axis=1)
cols = []
for i in range(len(sp500_companies.columns)):
    cols.append(sp500_companies.columns[i].lower())
sp500_companies.columns = cols
```

```python
len(sp500_companies['symbol'])  # number of companies
```

```python
sp500_quotes=quotes_all[quotes_all['symbol'].isin(sp500_companies['symbol'])]   # filtering quotes of S&P500 companies only
```

```python
sp500_quotes
```

```python
sp500_quotes['symbol'].nunique()  # 6 companies with quotes missed
```

```python
symbols_sp500=sp500_companies['symbol'].tolist()
symbols=(sp500_quotes['symbol'].unique()).tolist()
```

```python
missing_companies=[]
for i in range(len(symbols_sp500)):
    if symbols_sp500[i] not in symbols:
        missing_companies.append(symbols_sp500[i])
```

```python
missing_companies   # low weight in index (<0.04%), except for BRK.B (1.52%)
```

```python
sp500_companies[sp500_companies['symbol'].isin(missing_companies)]
```

```python
sp500_companies['sector'].value_counts()   # number of companies by sector
```

```python
sp500_weights= pd.read_csv('sp500_weights.csv')   # dataframe containing wieight of each company in index
```

```python
sp500_weights = sp500_weights.drop(['company'], axis=1)
```

```python
sp500_weights['symbol'].nunique()
```

```python
sp500_weights['weight'].sum()  # checking that sum of weights =100 (100%)
```

```python
symbols_sp500=sp500_companies['symbol'].tolist()
symbols=(sp500_weights['symbol'].unique()).tolist()
```

```python
missing_companies=[]
for i in range(len(symbols_sp500)):
    if symbols_sp500[i] not in symbols:
        missing_companies.append(symbols_sp500[i])
```

```python
missing_companies  # list of companies matches que previous list (sp500)
```

```python
import re
```

```python
sp500['datetime'] = sp500['datetime'].apply(lambda x :re.findall('\d{4}-\d{2}-\d{2}', x)[0])
```

```python
# creating a DataFrame with all the information 

sp500=pd.merge(left =   sp500_quotes,
                        right = sp500_companies,
                        how = 'inner', 
                        left_on = "symbol", 
                        right_on= "symbol")
```

```python
sp500
```

```python
sp500=pd.merge(left =   sp500,
                        right = sp500_weights,
                        how = 'inner', 
                        left_on = "symbol", 
                        right_on= "symbol")
```

```python
sp500
```

```python
sp500 = sp500.rename(columns={'datetime':'date'})
sp500 = sp500[['symbol','date','open_price','close_price','volume','company','sector','weight']]
```
